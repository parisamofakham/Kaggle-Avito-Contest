### Model building

1. **Data preprocessing** was carried out in several steps:
<br> a. Base preprocessing
<br> b. Keywords generation
<br> c. Image features
<br> d. Text polarity
<br> e. Geographic features
<br>
<br>
2. Various **Base models** were built using newly processed data
<br> a. Neural Network
<br> b. XGBoost
<br> c. Ridge (L1 regularization) 
<br> d. Recurrent Neural Network  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    * - note that (f) does not include the same data preprocessing as the other base models (all preprocessing is included in the same file)*
<br>
<br>
3. **Intermediary models** were created using both base models from (2) and select features from the preprocessed data
<br> a. Stacked LightGBM combining base models (a), (b), and (c) from (2) and select features from preprocessed data
<br> b. LightGBM using base model (c) from (2) and select features from preprocessed data
<br> c. Categorical LightGBM using base model (c) from (2) and select features from preprocessed data
<br>
<br>
4. Final **Blended model**
<br> a. Blended model using intermediary models (a), (b) and (c) from (3), and base models (b), (c), (d), (e), and (f) from (2)
